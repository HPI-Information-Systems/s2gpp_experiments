# Data Exploration

```python
import pandas as pd
import numpy as np
import plotly.offline as py
import plotly.graph_objects as go
import plotly.figure_factory as ff
import plotly.express as px
from plotly.subplots import make_subplots
from pathlib import Path
from sklearn.preprocessing import MinMaxScaler
from typing import Optional
```

```python
def plot_ts(dataset: pd.DataFrame, score: Optional[np.ndarray] = None):
    
    
    fig = go.Figure()
    
    for i in range(1, len(dataset.columns)):
        c = dataset.columns[i]
        line = go.Scatter(y=MinMaxScaler().fit_transform(dataset.loc[:, [c]]).reshape(-1), name=c)
        fig.add_trace(line)
    
    if score is not None:
        line = go.Scatter(y=MinMaxScaler().fit_transform(score.reshape(-1, 1)).reshape(-1), name="score")
        fig.add_trace(line)

    fig.show()
```

```python
def get_dataset(name: str, collection="GutenTAG", learntype="test") -> pd.DataFrame:
    data_dir = Path("../data/")
    filepath = data_dir / collection / name / (learntype + ".csv")
    return pd.read_csv(filepath)
```

```python
def get_score(path: Path) -> np.ndarray:
    return np.genfromtxt(path)
```

```python
plot_ts(get_dataset("sinus-combined-diff-1"), get_score(Path("../anomaly_scores.ts")))
```
